```python
import matplotlib.pyplot as plt
import numpy as np

# 配置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # 黑体，常见的中文字体
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

# 数据
categories = ['I2T_R@1', 'I2T_R@5', 'I2T_R@10', 'T2I_R@1', 'T2I_R@5', 'T2I_R@10', 'Mean_R@K']
Clip = [9.51, 23.01, 32.74, 8.81, 27.88, 43.19, 24.19]
RemoteClip = [27.88, 51.55, 63.27, 23.67, 59.38, 74.82, 50.1]
RemoteClipAndGpt_4o = [29.42, 56.19, 66.81, 26.73, 59.65, 75.09, 52.32]

# 设置条形宽度
bar_width = 0.25
index = np.arange(len(categories))

# 绘制条形图
fig, ax = plt.subplots(figsize=(10, 6))  # Adjusted figure size for better readability

bar1 = ax.bar(index, Clip, bar_width, label='Clip', color='royalblue')
bar2 = ax.bar(index + bar_width, RemoteClip, bar_width, label = 'RemoteClip', color = 'red')
bar3 = ax.bar(index + bar_width * 2, RemoteClipAndGpt_4o, bar_width, label='RemoteClipAndGpt_4o', color='darkorange')

# 添加百分比标注
for rect in bar1:
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width() / 2, height + 1, f'{height}%', ha='center', va='bottom', fontsize=10)

for rect in bar2:
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width() / 2, height, f'{height}%', ha='center', va='bottom', fontsize=10)

for rect in bar3:
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width() / 2, height + 1, f'{height}%', ha='center', va='bottom', fontsize=10)

# 添加标签和标题
ax.set_xlabel('不同的R@K')
ax.set_ylabel('不同的R@K对应的值')
ax.set_title('通用多模态对比模型与我们的方法的对比')
ax.set_xticks(index + bar_width)
ax.set_xticklabels(categories, rotation=45, ha='right')  # Rotate x-tick labels for better readability
ax.legend()

# 调整布局，避免标签重叠
plt.tight_layout()
plt.show()
```
