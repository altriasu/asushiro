# 1 张量是什么？
张量是一个多维数组，它是标量、向量、矩阵的高维拓展

torch.Tensor属性
- data: 被包装的Tensor
- grad：data的梯度
- grad_fn：创建Tensor的Function，是自动求导的关键
- requires_grad:指示是否需要梯度
- is_leaf：指示是否是叶子结点（张量）
- dtype：张量数据类型，如torch.FloatTensor，torch.cuda.FloatTensor
- shape：张量的形状，如（64,3,224,224）
- device：张量所在设备，GPU/CPU

## 1.2.2 切分
### - torch.chunk()
功能：将张量按维度dim进行平均切分

返回值：张量列表

注意：`若不能整除，最后一份张量小于其他张量`
- input: 要切分的张量
- chunks：要切分的份数
- dim：要切分的维度

```python
a = torch.arange(1, 11).reshape((2, 5))
list_of_tensor = torch.chunk(a, dim = 1, chunks = 2)

for idx, t in enumerate(list_of_tensor):
    print("第{}个张量:{}, shape is {}".format(idx + 1, t, t.shape))
```

### - torch.split()
功能：将张量按维度dim进行切分

返回值：张量列表

- tensor:要切分的张量
- split_size_or_sections: 为int时，表示每一份的长长度；为list时，表示按list元素切分
- dim：要切分的维度

```python
a = torch.arange(1, 11).reshape((2, 5))
list_of_tensor = torch.split(a, 2, dim = 1) #按列切

for idx, t in enumerate(list_of_tensor):
    print("第{}个张量:{}, shape is {}".format(idx + 1, t, t.shape))
```

```python
a = torch.arange(1, 11).reshape((2, 5))
list_of_tensor = torch.split(a, [2, 1, 2], dim = 1) #按列表索引切

for idx, t in enumerate(list_of_tensor):
    print("第{}个张量:{}, shape is {}".format(idx + 1, t, t.shape))
```

# 例：线性回归

求解步骤：
1. 确定模型
$Model: y = wx + b$
2. 选择损失函数
$MSE: \dfrac{1}{m} \sum\limits^{m}_{i = 1}(y_i - \hat{y_i})^2$
3. 求解梯度并更新w, b
    
    $w = w - LR*w.grad$
    
    $b = b -LR*w.grad$

```python
import torch
from matplotlib import pyplot as plt
import numpy as np

# 设置学习率
lr = 0.01

# 创建训练数据
x = torch.rand(20, 1) * 10  # 随机生成20个数据点，范围为[0, 10)
y = 2 * x + (5 + torch.randn(20, 1))  # 生成标签，y = 2x + 5 + 噪声

# 构建线性回归参数
w = torch.randn(1, requires_grad=True)  # 初始化权重w
b = torch.zeros(1, requires_grad=True)  # 初始化偏置b

# 训练模型
for iteration in range(1000):
    # 前向传播
    y_pred = w * x + b  # 计算预测值

    # 计算损失函数 (均方误差)
    loss = (0.5 * (y - y_pred) ** 2).mean()  # 向量运算，取平均

    # 反向传播
    loss.backward()

    # 更新参数
    w.data.sub_(lr * w.grad)
    b.data.sub_(lr * b.grad)


    # 每隔20次迭代，绘制当前拟合曲线
    if iteration % 20 == 0:
        plt.scatter(x.data.numpy(), y.data.numpy())  # 绘制原始数据点
        plt.plot(x.data.numpy(), y_pred.data.numpy(), 'r-', lw=5)  # 绘制拟合直线
        plt.text(2, 20, f'Loss={loss.item():.4f}', fontdict={'size': 20, 'color': 'red'})
        plt.ylim(8, 28)
        plt.xlim(1.5, 10)
        plt.title(f"Iteration: {iteration}\n w: {w.item():.4f} b: {b.item():.4f}")
        plt.pause(0.5)

        # 如果损失函数的值小于1，则停止迭代
        if loss.item() < 1:
            break
```

# 2 计算图
计算图是用来描述运算的有向无环图\
主要两个元素：节点（node）和边（edge）

节点表示`数据`，如向量，矩阵，张量\
边表示运算，如加减乘除卷积

例如用计算图表示：$y = (x + w ) * ( w + 1 )$\
$a = x + w$\
$b = w + 1$\
$y = a * b$\
<img src="https://cdn.acwing.com/media/article/image/2024/08/26/291897_d3ea115463-%E5%B1%8F%E5%B9%95%E6%88%AA%E5%9B%BE-2024-08-26-194612.png" style="width:500px; height:300px"/>

$\dfrac{\alpha y}{\alpha w} = \dfrac{\alpha y}{\alpha a} \dfrac{\alpha a}{\alpha w} + \dfrac{\alpha y}{\alpha b} \dfrac{\alpha b}{\alpha w}$\
$= b * 1 + a * 1$\
$= b + a$\
$= (w + 1) + (x + w)$\
$= 2 * w + x + 1$\
$= 2 * 1 + 2 + 1 = 5$

叶子节点：用户创建的节点称为叶子节点，如此题中的$x$ 与 $w$
is_leaf: 指示张量是否为叶子节点
 


```python
import torch

x = torch.tensor([2.], requires_grad = True)
w = torch.tensor([1.], requires_grad = True)

a = x + w
a.retain_grad() #加上保存缓冲后可以查到对应梯度
b = w + 1
y = a * b

y.backward()

print(w.grad)
```

```python
print("is_leaf:\n", w.is_leaf, x.is_leaf, a.is_leaf, b.is_leaf, y.is_leaf)
print("gradient:\n", w.grad, x.grad, a.grad, b.grad, y.grad) # 在反向传播之后, 非叶子节点的梯度都被释放掉了以节省内存开销
```

```python
print("grad_function:\n", w.grad_fn, x.grad_fn, a.grad_fn, b.grad_fn, y.grad_fn)
```

# 3 autograd--自动求导系统
### - torch.autograd.backward()
功能：自动求取梯度
- tensor：用于求导的张量，如loss
- retain_graph：保存计算图
- create_graph: 创建导数计算图，用于高阶求导
- grad_tensors：多梯度权重

### - torch.autograd.grad()
功能：求取梯度
- outputs: 用于求导的张量，如loss
- imputs：需要梯度的张量
- create_graph: 创建导数计算图，用于高阶求导
- retain_graph: 保存计算图
- grad_outputs: 多梯度权重

```python
# 例：求二阶导数
x = torch.tensor([3.], requires_grad = True)
y = torch.pow(x, 2) # y = x^2

grad_1 = torch.autograd.grad(y, x, create_graph = True)
print(grad_1)

grad_2 = torch.autograd.grad(grad_1[0], x)# grad_1是一个元组，取第一个tensor
print(grad_2)
```

### 注意
1. 梯度不自动清零
2. 依赖于叶子节点的节点requires_grad默认为True
3. 叶子结点不可执行in-place

# 例：逻辑回归
逻辑回归是线性的二分类模型\
模型表达式：\
$y = f(WX + b)$\
$f(x) = \dfrac{1}{1 + e^{-x}}$\
$f(x)$称为sigmoid函数，也称为Logistic函数

```python
import torch
import torch.nn as nn
from matplotlib import pyplot as plt
import numpy

# 生成数据
sample_nums = 100
mean_value = 1.7
bias = 1
n_data = torch.ones(sample_nums, 2) 
x_0 = torch.normal(mean_value * n_data, 1) + bias # 张量+标量 shape = (100, 2)
y_0 = torch.zeros(sample_nums) # shape = (100, 1) #zeros表示类别0
x_1 = torch.normal(-mean_value * n_data, 1) + bias
y_1 = torch.ones(sample_nums) # ones表示类别1

x_train = torch.cat([x_0, x_1], 0)
y_train = torch.cat([y_0, y_1], 0)

# 选择模型
class LR(nn.Module):
    def __init__(self):
        super(LR, self).__init__()
        self.features = nn.Linear(2, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x): 
        x = self.features(x)
        x = self.sigmoid(x)
        return x

lr_net = LR()# 实例化模型

# 选择损失函数
loss_fn = nn.BCELoss()#交叉熵损失函数

# 选择优化器
lr = 0.01
optimizer = torch.optim.SGD(lr_net.parameters(), lr = lr, momentum = 0.9)#选择SGD梯度下降法

for epoch in range(1000):
    #前向传播
    y_pred = lr_net(x_train)
    # 计算loss
    loss = loss_fn(y_pred.squeeze(), y_train)
    #反向传播
    loss.backward()
    #更新参数
    optimizer.step()
    #绘图
    if epoch % 20 == 0:
        mask = y_pred.ge(0.5).float().squeeze() # 以0.5为阈值分类
        correct = (mask == y_train).sum() #计算正确预测样本个数
        acc = correct.item() / y_train.size(0) #计算分类准确率

        plt.scatter(x_0.data.numpy()[:, 0], x_0.data.numpy()[:, 1], c = 'r', label = 'class 0')
        plt.scatter(x_1.data.numpy()[:, 0], x_1.data.numpy()[:, 1], c = 'b', label = 'class 1')

        w0, w1 = lr_net.features.weight[0]
        w0, w1 = float(w0.item()), float(w1.item())
        plot_b = float(lr_net.features.bias[0].item())
        plot_x = np.arange(-6, 6, 0.1)
        plot_y = (-w0 * plot_x - plot_b) / w1

        plt.xlim(-5, 7)
        plt.ylim(-7, 7)
        plt.plot(plot_x, plot_y)

        plt.text(-5, 5, 'Loss=%.4f' % loss.data.numpy(), fontdict={'size':20, 'color':'red'})
        plt.title("epoch: {}\nw0: {:.2f} w1: {:.2f} b:{:.2f} accuracy: {:.2%}".format(epoch, w0, w1, plot_b, acc))
        plt.legend()

        plt.show()
        plt.pause(0.5)

        if acc > 0.98999:
            break
```
